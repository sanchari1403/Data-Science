CATEGORICAL VARIABLE TREATMENT

```python
#conversion to binary variable
varlist =  ['aspiration', 'fueltype', 'doornumber', 'enginelocation']

# Defining the map function
def binary_map(x):
    return x.map({'std': 1, "turbo": 0,"gas": 1,"diesel": 0,"four": 1,"two": 0,"front": 1,"rear": 0})

# Applying the function to the housing list
carprice[varlist] = carprice[varlist].apply(binary_map)
```

```python
#Dummy variables created
status = pd.get_dummies(carprice['carbody'], drop_first = True)
carprice = pd.concat([carprice, status], axis = 1)
carprice.drop(['carbody'], axis = 1, inplace = True)
carprice.head()
```

```python
#Dummy variables created
status = pd.get_dummies(carprice['enginetype'], drop_first = True)
carprice = pd.concat([carprice, status], axis = 1)
carprice.drop(['enginetype'], axis = 1, inplace = True)
carprice.head()
```

```python
#Dummy variables created
status = pd.get_dummies(carprice['cylindernumber'], drop_first = True)
carprice = pd.concat([carprice, status], axis = 1)
carprice.drop(['cylindernumber'], axis = 1, inplace = True)
carprice.head()
```

```python
#Dummy variables created
status = pd.get_dummies(carprice['CarCompany'], drop_first = True)
carprice = pd.concat([carprice, status], axis = 1)
carprice.drop(['CarCompany'], axis = 1, inplace = True)
carprice.drop(['CarName'], axis = 1, inplace = True)
carprice.head()
```

```python
#Dummy variables created
status = pd.get_dummies(carprice['fuelsystem'], drop_first = True)
carprice = pd.concat([carprice, status], axis = 1)
carprice.drop(['fuelsystem'], axis = 1, inplace = True)
carprice.head()
```

SPLITTING OF DATA

```python
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
df_train, df_test = train_test_split(carprice, train_size = 0.7, test_size = 0.3, random_state = 100)
```

SCALING OF DATA

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
# Apply scaler() to all the columns except the 'yes-no' and 'dummy' variables
num_vars = ['car_ID','symboling', 'wheelbase', 'carlength', 'carwidth', 'carheight','curbweight','enginesize','boreratio','stroke','compressionratio','horsepower','peakrpm','citympg','highwaympg','carvolume','price']

df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
```

BUILDING A LINEAR MODEL

```python
# Importing RFE and LinearRegression
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
y_train = df_train.pop('price')
X_train = df_train
```

```python
#creating rfe using 25 columns
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm,25)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
#extracting those having rfe.support_ as True
col = X_train.columns[rfe.support_]
col
```

```python
#removing those having rfe.support_ as False
X_train.columns[~rfe.support_]
```

```python
X_train_rfe = X_train[col]
```

```python
#builduing the first model
import statsmodels.api as sm  
X_train_rfe = sm.add_constant(X_train_rfe)
lm = sm.OLS(y_train,X_train_rfe).fit() 
print(lm.summary())
```
