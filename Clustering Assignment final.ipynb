# Finding Principal Components

```python
#Improting the PCA module
from sklearn.decomposition import PCA
pca=PCA(0.93,random_state=42)
```

```python
#Doing the PCA on the original data
country1=country['country']
del country['country']
```

```python
country.shape
```

```python
#Scaling by normalisation 
normalized_country=(country-country.mean())/country.std()
country=country.drop(['child_mort', 'exports', 'health', 'imports', 'income', 'inflation','life_expec', 'total_fer', 'gdpp'],1)
country = pd.concat([country,normalized_country],axis=1)
country.head()
```

```python
#Doing the PCA on the original data
pca=pca.fit(country)
```

```python
pca.n_components_
```

```python
colnames = list(country.columns)
pcs_df = pd.DataFrame({'PC1':pca.components_[0],'PC2':pca.components_[1], 'Feature':colnames})
pcs_df.head()
```

```python
%matplotlib inline
fig = plt.figure(figsize = (8,8))
plt.scatter(pcs_df.PC1, pcs_df.PC2)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
for i, txt in enumerate(pcs_df.Feature):
    plt.annotate(txt, (pcs_df.PC1[i],pcs_df.PC2[i]))
plt.tight_layout()
plt.show()
```

```python
pca.explained_variance_ratio_
```

```python
#Making the screeplot - plotting the cumulative variance against the number of components
%matplotlib inline
fig = plt.figure(figsize = (12,8))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```

```python
#defining Hopkins
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
# Kmeans with K=5
model_clus5 = KMeans(n_clusters = 5, max_iter=50,random_state=42)
model_clus5.fit(x_pca)
```

```python
from sklearn.metrics import silhouette_score
sse_ = []
for k in range(2, 15):
    kmeans = KMeans(n_clusters=k).fit(x_pca)
    sse_.append([k, silhouette_score(x_pca, kmeans.labels_)])
```

```python
plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1]);
```

```python
# sum of squared distances
ssd = []
for num_clusters in list(range(1,21)):
    model_clus = KMeans(n_clusters = num_clusters, max_iter=50)
    model_clus.fit(x_pca)
    ssd.append(model_clus.inertia_)

plt.plot(ssd)
```
