```python
# Import the numpy and pandas packages

import numpy as np
import pandas as pd
```

```python
#reading the csv file into the dataframe

uber=pd.read_csv(r"E:\Data Science\Uber Request Data.csv")
uber
```

```python
#Data Cleaning-Modifying the Timestamps to appropriate format

uber["Request timestamp"]=pd.to_datetime(uber["Request timestamp"],dayfirst=True)
uber["Drop timestamp"]=pd.to_datetime(uber["Drop timestamp"],dayfirst=True)
uber
```

```python
#extracting Date and Time from the Request timestamp

uber["Request Date"]=uber["Request timestamp"].apply(lambda x:str(x).split()[0])
uber["Request Time"]=uber["Request timestamp"].apply(lambda x:str(x).split()[1])
uber
```

```python
#extracting hours from Request Time

uber["Hours"]=uber["Request Time"].apply(lambda x:x.split(':')[0])
uber["Hours"]=uber["Hours"].apply(lambda x:int(x))
uber
```

```python
#Segregating the data into different timeslots based on request hour

uber.loc[(uber["Hours"]>=4) & (uber["Hours"]<=8),'Timeslot']="Early Morning"
uber.loc[(uber["Hours"]>8) & (uber["Hours"]<=11),'Timeslot']="Morning"
uber.loc[(uber["Hours"]>11) & (uber["Hours"]<=15),'Timeslot']="Afternoon"
uber.loc[(uber["Hours"]>15) & (uber["Hours"]<=19),'Timeslot']="Evening"
uber.loc[(uber["Hours"]>19) & (uber["Hours"]<=23),'Timeslot']="Night"
uber.loc[(uber["Hours"]>=0) & (uber["Hours"]<4),'Timeslot']="Late Night"
uber
```

```python
#grouping the data by STatus, Timeslot, Pickup point

div1=uber.groupby(["Status",'Timeslot','Pickup point'])
div1=pd.DataFrame(div1["Timeslot"].count().rename('Count').reset_index())

#defining a derived column for analysis
div1["X-Axis"]=''
for i in range(div1["Status"].count()):
    if (div1["Status"][i]=='Cancelled') & (div1["Pickup point"][i]=='Airport'):
        div1["X-Axis"][i]='Cancelled-Airport'
    elif (div1["Status"][i]=='Cancelled') & (div1["Pickup point"][i]=='City'):
        div1["X-Axis"][i]='Cancelled-City'
    elif (div1["Status"][i]=='No Cars Available') & (div1["Pickup point"][i]=='City'):
        div1["X-Axis"][i]='No Cars-City'
    elif (div1["Status"][i]=='No Cars Available') & (div1["Pickup point"][i]=='Airport'):
        div1["X-Axis"][i]='No Car-Airport'
    elif (div1["Status"][i]=='Trip Completed') & (div1["Pickup point"][i]=='City'):
        div1["X-Axis"][i]='Trip Completed-City'
    elif (div1["Status"][i]=='Trip Completed') & (div1["Pickup point"][i]=='Airport'):
        div1["X-Axis"][i]='Trip Completed-Airport'
    else:
        div1["X-Axis"][i]=''
div1
```

```python
#Graph showing the frequency of requests that get cancelled or show 'no cars available'
#identifying the most problematic types of requests (city to airport / airport to city etc.)
#identifying the time slots (early mornings, late evenings etc.)

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
plt.figure(figsize=(12, 6))

sns_plt=sns.barplot(x="X-Axis", y="Count", hue="Timeslot", data=div1);
plt.title("Status-Time Slot")

plt.show()
fig=sns_plt.get_figure()
fig.savefig('Res_1.jpeg')
```

```python
#Defining a derived column from a column - Univariate analysis

uber['SrcDest']=''
for i in range(uber["Status"].count()):
    if uber["Pickup point"][i]=='Airport':
        uber['SrcDest'][i]='Airport-City'
    else:
        uber['SrcDest'][i]='City-Airport'
uber
```

```python
#Defining a derived column from a column - Univariate analysis

uber["Supply_Stat"]=1
for i in range(uber["Status"].count()):
    if uber["Status"][i]=="Trip Completed":
        uber["Supply_Stat"][i]=1
    else:
        uber["Supply_Stat"][i]=0
uber
```

```python
#Grouping the data

timeslot1=uber.groupby("Hours")
timeslot2=uber.groupby(['Hours',"SrcDest"])
```

```python
#Finding the gap in supply and demand with respect to Hours

timeslot1_Sup=pd.DataFrame(timeslot1["Supply_Stat"].sum().rename('Supply').reset_index())
timeslot1_Dem=pd.DataFrame(timeslot1["Request id"].count().rename('Demand').reset_index())
gap1=timeslot1_Sup.merge(timeslot1_Dem)
gap1["Gap"]=gap1["Demand"]-gap1["Supply"]
gap1
```

```python
#Finding the time slots when the highest gap exists


import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="white")

plt.plot(gap1['Hours'],gap1['Supply'])
plt.plot(gap1['Hours'],gap1['Demand'])
plt.plot(gap1['Hours'],gap1['Gap'])
plt.xlabel("Time Slot")
plt.ylabel("Request Frequency")
plt.title("Supply-Demand vs Time Slot")
plt.legend()

plt.savefig('res2.jpeg')
```

```python
#Finding the gap in supply and demand with respect to Hours and Pickup point

timeslot2_Sup=pd.DataFrame(timeslot2["Supply_Stat"].sum().rename('Supply').reset_index())
timeslot2_Dem=pd.DataFrame(timeslot2["Request id"].count().rename('Demand').reset_index())
gap2=timeslot2_Sup.merge(timeslot2_Dem)
gap2["Gap"]=gap2["Demand"]-gap2["Supply"]
gap2
```

```python
#Finding the types of requests (city-airport or airport-city) for which the gap is the most severe in the identified time slots

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

sns_plt=sns.barplot(x="Hours", y="Gap", hue="SrcDest", data=gap2);
plt.title("Gap-Time")

plt.show()
fig=sns_plt.get_figure()
fig.savefig('Res_3.jpeg')
```

```python
#Finding day of the week and counting the total requests on each day

uber['DayOfWeek'] = uber['Request timestamp'].dt.weekday_name
div2=uber.groupby("DayOfWeek")["Request id"].count().rename('Demand').reset_index()
div2.sort_values(by='Demand',ascending=False)
```

```python
#Finding the shortage in cars on each day


uber_byWeekDay=uber[((uber["Status"]=='No Cars Available') | (uber["Status"]=='Cancelled'))]
div2_1=uber_byWeekDay.groupby('DayOfWeek')["Request id"].count().rename('Gap').reset_index()
div2=div2.merge(div2_1)
div2
```

```python
#Finding the week days when the highest gap exists


import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="white")

plt.plot(div2['DayOfWeek'],div2['Demand'])
plt.plot(div2['DayOfWeek'],div2['Gap'])

plt.xlabel("Day Of Week")
plt.ylabel("Request Frequency")
plt.title("Supply-Demand vs Day")
plt.legend()

plt.savefig('res4.jpeg')
```

```python
#Finding the gap with repect to Source, Destination and Day of the Week

div3=uber.groupby(["DayOfWeek","SrcDest"])["Request id"].count().rename('Demand').reset_index()
div3_1=uber[((uber["Status"]=='No Cars Available') | (uber["Status"]=='Cancelled'))].groupby(["DayOfWeek","SrcDest"])["Request id"].count().rename('Gap').reset_index()
div3=div3.merge(div3_1)
div3
```

```python
#Finding the types of requests (city-airport or airport-city) for which the gap is the most severe in the identified weekdays

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

sns_plt=sns.barplot(x="DayOfWeek", y="Gap", hue="SrcDest", data=div3);
plt.title("Gap-Time")

plt.show()
fig=sns_plt.get_figure()
fig.savefig('Res_5.jpeg')
```
